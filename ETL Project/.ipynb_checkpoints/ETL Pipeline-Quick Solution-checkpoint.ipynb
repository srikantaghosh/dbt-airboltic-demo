```python
s3 = boto3.resource('s3')
bucket = s3.Bucket('xetra-1234')

bucket_obj = bucket.objects.filter(Prefix='2022-01-28/')
```

```python
objects = [obj for obj in bucket_obj]

objects
```

```python
## Converting to Dataframe 
data = StringIO(csv_obj)

df= pd.read_csv(data, delimiter= ',')
df.head(50)
```

## Read Multiple Files

```python
bucket_obj1 = bucket.objects.filter(Prefix='2022-01-28/')
bucket_obj2 = bucket.objects.filter(Prefix='2022-01-31/')
objects = [obj for obj in bucket_obj1] + [obj for obj in bucket_obj2]

objects
```

```python
csv_obj_init = bucket.Object(key = objects[0].key).get().get('Body').read().decode('utf-8')
data = StringIO(csv_obj_init)
df_init = pd.read_csv(data, delimiter= ',')

df_init.columns
    
    
```

```python
df_all = pd.DataFrame(columns=df_init.columns)

# Use pd.concat to combine DataFrames
all_dfs = []  # Create a list to store DataFrames

for obj in objects:
    csv_obj = bucket.Object(key=obj.key).get().get('Body').read().decode('utf-8')
    data = StringIO(csv_obj)
    df = pd.read_csv(data, delimiter=',')
    all_dfs.append(df)  # Append the current DataFrame to the list

df_all = pd.concat(all_dfs, ignore_index=True)
```

```python
df_all
```

```python
columns = ['ISIN', 'Date', 'Time', 'StartPrice', 'MaxPrice', 'MinPrice',
       'EndPrice', 'TradedVolume']

df_all = df_all.loc[:, columns]
df_all
```

```python
## Filtering- Dropping cases where there is no data

df_all.dropna(inplace=True)
```
